## Question 4
Bivariate Analysis

```python
plt.plot(df["Age"],df["Fare"])         ##plot 9
plt.ylabel("Age")
plt.xlabel("Fare")
plt.show()
```

```python
sns.jointplot(df["Age"],df["Fare"],kind="hex",color="g")       ##plot 10
```

```python
sns.barplot(df['Age'],df['Fare'],hue='Sex',data=df,color='green')        ##plot 11
```

```python
sns.stripplot(x="Pclass",y="Fare",data=df)       ##plot 12
```

```python
sns.lmplot("Fare","Age",data=df,hue="Sex")      ##plot 13
```

```python
sns.lmplot("Fare","Age",data=df,hue="Sex",fit_reg=False)      ##plot 14
```

```python
sns.swarmplot(df["Pclass"],df["Fare"],data=df)        ##plot 15
```

```python
sns.kdeplot(df["Age"],df["Fare"])    ##plot 16
```

```python
plt.scatter(df["Age"],df["Fare"],c="g",marker="*")     ##plot 17
plt.ylabel("Age")
plt.xlabel("Fare")
plt.show()
```
